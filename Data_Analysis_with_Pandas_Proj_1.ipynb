```python
import pandas as pd

dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,99],
                'Gender':['Male','Female','Female','Male','Male',
                         'Female','Female']
               }
df1=pd.DataFrame(dict1)
df1
```

```python
# Display top 3 Rows of the dataset

df1.head(3)
```

```python
# Check last 3 rows of the dataset

df1.tail(3)
```

```python
# Find shape of our dataset (number of rows & number of columns)
# shape is a attribute not a fn
df1.shape

print("Number of rows",df1.shape[0])
print("Number of columns",df1.shape[1])
```

```python
# Get info about out dataset like total number of rows, total number of columns, datatypes of each column and memory requirement

df1.info()
```

```python
# check null values in dataset

# df1.isnull().sum(axis=1) # axis = 1 to check rows wise
df1.isnull().sum(axis=0)# axis = 0 to check column wise
```

```python
# get overall statistics about the dataframe

df1.describe() # by default this method only picks the numerical columns
# to include character column, pass include parameter inside describe(include ='all') fn.

df1.describe(include ='all')
```

```python
# find unique values from gender column

df1['Gender'].unique()
```

```python
# find the number of unique values from the gender column
df1['Gender'].nunique()
```

```python
# find the count of unique values in gender column
# df1.groupby('Gender').count()
df1["Gender"].value_counts()
```

```python
# find the total number of students having marks between 90 to 100(inclusive) using between method
# method 1
# len(df1[(df1["Marks"]>=90) & (df1["Marks"]<=100)])

# method 2
sum(df1['Marks'].between(90,100))
```

```python
# find average marks

round(df1['Marks'].mean(),2)
```

```python
# apply method: To use user defined functions & inbuilt python funcation
def marks(x):
  return x/2

df1['Marks'].apply(marks) #call function without paranthesis
```

```python
# lambda fn with apply fn

df1['Marks'].apply(lambda x:x/2)
```

```python
# map fn

df1["Male_Female_Ind"]=df1["Gender"].map({"Male":1, "Female": 0})

df1
```

```python
# Drop the column(s) -- for multiple column drop use list.
# inpace = True is used to modify the original dataset
df1.drop("Male_Female_Ind",axis = 1, inplace = True)
```

```python
# sort the dataframe as per the marks columns
# for multi column sorting use list
df1.sort_values(by="Marks", ascending=False)
```
